## Indices Distribution

```python
def format_ax(ax):
    ax.minorticks_on()
    ax.xaxis.set_major_locator(MultipleLocator(.2))
    ax.yaxis.set_major_locator(MultipleLocator(.2))
    ax.xaxis.set_minor_locator(MultipleLocator(.1))
    ax.yaxis.set_minor_locator(MultipleLocator(.1))
    ax.grid(which='both', axis='y', linestyle='-', linewidth=.3, color='lightgrey')
    ax.set_ylim((0,1))
    ax.set_xlim((0,1))
    ax.spines['left'].set_position(('outward', 5))
    ax.spines['bottom'].set_position(('outward', 2))
    return ax
```

```python
bins = np.linspace(0, 1, 11)
%matplotlib inline
fig, axes = plt.subplots(1, 3, figsize=(14/2.54, 5/2.54), dpi=300, sharey=True, sharex=True)
sns.histplot(df_llmgadm_gdisgadm['jaccard'], ax=axes[0], bins=bins, stat='probability', alpha=0.7, color='lightgrey',edgecolor=None, element = "step")
axes[0].set_title("LLM GADM vs. EM-DAT GAUL")
sns.histplot(df_llmgadm_emdatgaul['jaccard'], ax=axes[1], bins=bins, stat='probability', alpha=0.7,color='lightgrey', edgecolor=None, element = "step")
axes[1].set_title("LLM GADM vs. GDIS GADM")
sns.histplot(df_llmgadm_emdatgaul['jaccard'], ax=axes[2], bins=bins, stat='probability', alpha=0.7,color='lightgrey', edgecolor=None, element = "step")
axes[2].set_title("EM-DAT GADM vs. GDIS GADM")

axes[0].set_ylim((-0.2,1.2))

for ax in axes.ravel():
    format_ax(ax)
fig.tight_layout()
```
